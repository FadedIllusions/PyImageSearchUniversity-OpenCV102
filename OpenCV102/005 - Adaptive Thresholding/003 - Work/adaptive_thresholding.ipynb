```python
# Import Needed Packages
import matplotlib.pyplot as plt
import cv2
```

```python
def plt_imshow(title, image):
    # Convert From BGR To RGB And Display
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
# # construct the argument parser and parse the arguments
# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--image", type=str, required=True,
# help="path to input image")
# args = vars(ap.parse_args())

# since we are using Jupyter Notebooks we can replace our argument
# parsing code with *hard coded* arguments and values
args = {
    "image": "images/steve_jobs.png"
}
```

```python
# Load Image And Display
image = cv2.imread(args["image"])
plt_imshow("Image", image)
```

```python
# Convert To Grayscale And Blur
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (7,7), 0)
```

```python
# Apply Simple Thresholding With Hardcoded Threshold Value
(T, threshInv) = cv2.threshold(blurred, 230, 255, 
                               cv2.THRESH_BINARY_INV)
plt_imshow("Simple Thresholding", threshInv)
```

```python
# Apply Otsu's Automatic Thresholding
(T, treshInv) = cv2.threshold(blurred, 0, 255,
                cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
plt_imshow("Otsu Thresholding", threshInv)
```

```python
# Use Adaptive Thresholding
# (Examines Neighboring Pixels And Adaptively Thresholds Each)
thresh = cv2.adaptiveThreshold(blurred, 255, 
        cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 21, 10)
plt_imshow("Mean Adaptive Thresholding", thresh)
```

```python
# Use Adaptive Thresholding Using Gaussian Weighting
thresh = cv2.adaptiveThreshold(blurred, 255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 21, 4)
plt_imshow("Guassian Adaptive Thresholding", thresh)
```
